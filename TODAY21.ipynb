```python
# Function to get a list of integers from user input
def get_list_from_input(prompt,n):
  while True:
    user_input = input(prompt)
    substrings = user_input.split()
    if len(substrings) == n:
      try:
         numbers = list(map(int, substrings))
         return numbers
      except ValueError:
          print("Please enter valid integers separated by spaces.")
    else:
        print(f"Please enter exactly {n} numbers.")
n=int(input("Enter the range of a list "))
list1 = get_list_from_input("Enter the first list of numbers ",n)
list2 = get_list_from_input("Enter the second list of numbers ",n)
combined_list = list1 + list2
unique_values = set(combined_list)
result_list = list(unique_values)
print("The list without duplicates is:", result_list)
```

```python
# List Sorting Without sort(): Implement bubble sort on a list
def bubble_sort(arr):
  n = len(arr)
  for i in range(n):
    for j in range(0, n-i-1):
      if arr[j] > arr[j+1]:
        arr[j],arr[j+1]=arr[j+1],arr[j]
if __name__=="__main__":
    Given_list = [64, 34, 25, 12, 22, 11, 90]
    print("Original list ",Given_list)
    bubble_sort(Given_list)
    print("Sorted list ",Given_list)
```

```python
# Find Second Largest Number in a List: Without using max() function.
def find_second_largest(arr):
  if len(arr) < 2:
    return None
  first = second = float('-inf')
  for number in arr:
      if number > first:
        second = first
        first = number
      elif first > number > second:
          second = number
  return second if second != float('-inf') else None
if __name__ == "__main__":
    Given_list = [64, 34, 25, 12, 22, 11, 90]
    print("Second largest number is ", find_second_largest(Given_list))
```

```python
# Merge Two Sorted Lists: Combine and sort two lists
def merge_sorted_lists(list1, list2):
    merged_list = []
    i, j = 0, 0
    while i < len(list1) and j < len(list2):
        if list1[i] < list2[j]:
            merged_list.append(list1[i])
            i += 1
        else:
            merged_list.append(list2[j])
            j += 1
    merged_list.extend(list1[i:])  # Append remaining elements from list1
    merged_list.extend(list2[j:])  # Append remaining elements from list2
    return merged_list
if __name__ == "__main__":
    list1 = [1, 3, 5, 7]
    list2 = [2, 4, 6, 8]
    merged = merge_sorted_lists(list1, list2)
    print("Merged sorted list ", merged)
```

```python
# Find Common Elements in Two Lists: Using loops and set operations
# using loop
def find_common_elements_loops(list1, list2):
    common_elements = []
    for element in list1:
        if element in list2 and element not in common_elements:
            common_elements.append(element)
    return common_elements
if __name__ == "__main__":
    list1 = [1, 2, 3, 4, 5]
    list2 = [4, 5, 6, 7, 8]
    common = find_common_elements_loops(list1, list2)
    print("Common elements ", common)
```

```python
# Using set operations
def find_common_elements_sets(list1, list2):
    return list(set(list1) & set(list2))
if __name__ == "__main__":
    list1 = [1, 2, 3, 5, 9]
    list2 = [2, 3, 6, 7, 4]
    common = find_common_elements_sets(list1, list2)
    print("Common elements ", common)
```

```python
# Data Hiding Example: Implement private variables and access them through methods
class BankAccount:
    def __init__(self, account_number, balance=0):
        self.__account_number = account_number
        self.__balance = balance
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited {amount:.2f}")
        else:
            print("Deposit amount must be positive")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdraw {amount:.2f}")
        else:
            print("Invalid withdrawal amount")

    def get_balance(self):
        return self.__balance

    def get_account_number(self):
        return self.__account_number

# Example usage
if __name__ == "__main__":
    account = BankAccount("123456789")
    account.deposit(1000)
    account.withdraw(500)
    print("Account Number is ", account.get_account_number())
    print("Current Balance  ", account.get_balance())
```

```python
# Student Grading System Using Classes: Create a class to assign grades.
class Student:
    def __init__(self, name):
        self.name = name
        self.marks = []

    def add_marks(self, marks):
        if 0 <= marks <= 100:
            self.marks.append(marks)
        else:
            print("Marks must be between 0 and 100")

    def calculate_average(self):
        return sum(self.marks) / len(self.marks) if self.marks else 0

    def assign_grade(self):
        average = self.calculate_average()
        if average >= 90:
            return 'A'
        elif average >= 80:
            return 'B'
        elif average >= 70:
            return 'C'
        elif average >= 60:
            return 'D'
        return 'F'

    def display_info(self):
        average = self.calculate_average()
        grade = self.assign_grade()
        print(f"Student Name is {self.name}")
        print(f"Marks  {self.marks}")
        print(f"Average marks  {average:.2f}")
        print(f"Grade  {grade}")

# Example usage
if __name__ == "__main__":
    student = Student("Advik")
    student.add_marks(85)
    student.add_marks(92)
    student.add_marks(78)
    student.add_marks(88)
    student.display_info()
```

```python
# SQLite Database Connection: Connect Python to an SQLite database.
import sqlite3
# Connect to SQLite database (or create it if it doesn't exist)
connection = sqlite3.connect('students.db')
cursor = connection.cursor()
# Create a table
cursor.execute('''
CREATE TABLE IF NOT EXISTS students (
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT NOT NULL,
    age INTEGER NOT NULL,
    grade TEXT NOT NULL
)
''')
# Function to insert a new student
def insert_student(name, age, grade):
    cursor.execute('''
    INSERT INTO students (name, age, grade) VALUES (?, ?, ?)
    ''', (name, age, grade))
    connection.commit()
# Function to fetch all students
def fetch_students():
    cursor.execute('SELECT * FROM students')
    return cursor.fetchall()
# Example usage
if __name__ == "__main__":
    insert_student('Alice', 20, 'A')
    insert_student('Bob', 21, 'B')
    insert_student('Charlie', 22, 'C')

    students = fetch_students()
    print("Students in the database:")
    for student in students:
        print(f"ID: {student[0]}, Name: {student[1]}, Age: {student[2]}, Grade: {student[3]}")
# Close the connection
connection.close()
```

```python
# Multiple Exception Handling: Handle ValueError and ZeroDivisionError
def divide_numbers():
    try:
        n = float(input("Enter the first number "))
        d = float(input("Enter the second number "))
        re = n / d
        print(f"The result of {n} divided by {d} is {re:.2f}")
    except ValueError:
        print("Invalid input! Again Enter")
    except ZeroDivisionError:
        print("Error! Not Division by zero ")

if __name__ == "__main__":
    divide_numbers()
```

```python
# Try-Finally Block: Demonstrate resource cleanup using finally.
def write_to_file(filename, data):
    try:
        file = open(filename, 'w')
        file.write(data)
        print("Data written to file successfully")
    finally:
        file.close()
        print("File closed")
if __name__ == "__main__":
    write_to_file('example.txt', 'Hello, World!')
```

```python
# User-Defined Exception: Create a custom exception for age validation
class AgeValidationError(Exception):
    def __init__(self, age):
        self.age = age
        super().__init__(f"Invalid age {age}. Age must be between 0 and 120.")
def validate_age(age):
    if age < 0 or age > 120:
        raise AgeValidationError(age)
    print(f"Age {age} is valid.")
if __name__ == "__main__":
    try:
        age_input = int(input("Enter your age "))
        validate_age(age_input)
    except ValueError:
        print("Invalid age Value")
    except AgeValidationError as e:
        print(e)
```

```python
# Write a socket program for client-server communication //server side
import socket
def start_server():
    server_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    server_socket.bind(('localhost', 65432))
    server_socket.listen()
    print("Server is listening for connections...")
    conn, addr = server_socket.accept()
    with conn:
        print(f"Connected by {addr}")
        while True:
            data = conn.recv(1024)
            if not data:
                break
            print(f"Received: {data.decode()}")
            conn.sendall(data)  # Echo back the received data
if __name__ == "__main__":
    start_server()
```

```python
# client side
import socket
def start_client():
    client_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    client_socket.connect(('localhost', 65432))
    message = input("Enter a message to send to the server: ")
    client_socket.sendall(message.encode())
    data = client_socket.recv(1024)
    print(f"Received from server: {data.decode()}")
    client_socket.close()
if __name__ == "__main__":
    start_client()
```

```python
# todo.py - Basic To-Do List App
tasks = []  # List to store tasks
def show_tasks():
    if not tasks:
        print("\nNo tasks yet!")
    else:
        print("\nYour Tasks:")
        for i, task in enumerate(tasks, 1):
            print(f"{i}. {task}")
def add_task():
    task = input("Enter a new task: ")
    tasks.append(task)
    print(f'"{task}" added to the list!')
def remove_task():
    show_tasks()
    try:
        task_num = int(input("\nEnter task number to remove: "))
        removed = tasks.pop(task_num - 1)
        print(f'"{removed}" removed from the list!')
    except (IndexError, ValueError):
        print("Invalid choice!")
def menu():
    while True:
        print("\nTo-Do List Menu:")
        print("1. Show Tasks")
        print("2. Add Task")
        print("3. Remove Task")
        print("4. Exit")
        choice = input("Choose an option (1-4): ")
        if choice == "1":
            show_tasks()
        elif choice == "2":
            add_task()
        elif choice == "3":
            remove_task()
        elif choice == "4":
            print("Exiting To-Do List App. Have a great day!")
            break
        else:
            print("Invalid choice! Try again.")
# Start the application
menu()
```
