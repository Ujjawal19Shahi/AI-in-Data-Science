# **Data Visulization With Matplotlib**

```python
# import library
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x=["Ram","Shyam","Mayank","Shekhar","Golu"]
y=[95,88,99,86,89]
plt.figure(figsize=(5,5))


plt.plot(x,y,marker="o",linestyle=":",linewidth=2,color="r")
plt.xlabel("Name of Student")
plt.ylabel("Marks")
plt.title("Marks of Student")
plt.grid(True)
plt.show()
```

```python
data={
    "marks":[55,88,89,78,89]
}
df=pd.DataFrame(data)
df.index = [5,10,15,20,25] # Set the index after creating the DataFrame

plt.plot(df["marks"],marker="o",linestyle=":",linewidth=2,color="b")
plt.xlabel("X-Axis")
plt.ylabel("Marks")
plt.title("Marks of Student")
plt.grid(True)
plt.show()
```

```python
plt.hist(x)
plt.show()
plt.plot(df["marks"],marker="o",linestyle=":",color="b")
plt.xlabel("X-Axis")
plt.ylabel("Marks")
plt.title("Marks of Student")


#plt.grid(True, axis='x', color='gray', linestyle='--', linewidth=1)
```

```python
plt.hist(df["marks"],bins=5,color="y",edgecolor="b",alpha=0.07)
plt.xlabel("marks")
plt.ylabel("Frequency")
plt.show()
```

PIE CHART

**BAR CHART**

```python
name=["Ram","Shyam","Mayank","Shekhar","Golu"]
marks=[95,88,99,86,89]
plt.figure(figsize=(5,5))
plt.bar(name,marks,color=["red","blue","green","yellow","black"], width=0.7)
for i in range(len(name)):
      plt.text(name[i], marks[i], f"{marks[i]}", ha='center', va='bottom')
plt.xlabel("Name")
plt.ylabel("Marks")
plt.legend(marks)
plt.grid(True)
plt.show()
```

```python
Age=[70,20,50,60,90]
Salary=[10000,20000,30000,40000,50000]
bubble_size=[200,300,400,500,600]
color=["red","blue","yellow","green","black"]
plt.scatter(Age,Salary,bubble_size,color,alpha=0.35,edgecolors="red")
for i in range(len(Age)):
      plt.text(Age[i], Salary[i], f"{Salary[i]}")
plt.colorbar(label="SHAHI")
plt.title("Age and Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.grid(True)
plt.show()
```
