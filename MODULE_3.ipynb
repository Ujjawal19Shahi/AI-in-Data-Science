# **Data Visulization With Matplotlib**

```python
# import library
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from matplotlib.lines import lineStyles
x=["Ram","Shyam","Mayank","Shekhar","Golu"]
y=[95,88,99,86,89]
plt.figure(figsize=(5,5))


plt.plot(x,y,marker="o",linestyle=":",linewidth=2,color="r")
plt.xlabel("Name of Student")
plt.ylabel("Marks")
plt.title("Marks of Student")
plt.grid(True)
plt.show()
```

```python
from numpy._core.defchararray import index
from matplotlib import lines
data={
    "marks":[55,88,89,78,89]
}
df=pd.DataFrame(data)
df.index = [5,10,15,20,25] # Set the index after creating the DataFrame

plt.plot(df["marks"],marker="o",linestyle=":",linewidth=2,color="b")
plt.xlabel("X-Axis")
plt.ylabel("Marks")
plt.title("Marks of Student")
plt.grid(True)
plt.show()
```

```python
plt.hist(x)
plt.show()
plt.plot(df["marks"],marker="o",linestyle=":",color="b")
plt.xlabel("X-Axis")
plt.ylabel("Marks")
plt.title("Marks of Student")


#plt.grid(True, axis='x', color='gray', linestyle='--', linewidth=1)
```

```python
plt.hist(df["marks"],bins=5,color="y",edgecolor="b",alpha=0.07)
plt.xlabel("marks")
plt.ylabel("Frequency")
plt.show()
```
